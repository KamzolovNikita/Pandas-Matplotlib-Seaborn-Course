# Создание полотна

**Сущность matplotlib.figure.Figure** - Контейнер верхнего уровня для всех элементов графика.

**Сущность matplotlib.axes.Axes** - Класс Axes представляет собой один подграфик на рисунке. Он содержит нанесенные на график данные, отметки осей, лейблы, заголовок, легенду и т.д. Его методы являются основным интерфейсом для управления отрисовки графика.

Метод figure - создание полотна
* num - Уникальный идентификатор для рисунка.
* figsize - Ширина, высота в дюймах (передается кортежем из двух чисел) (по умолчанию (6.4, 4.8))
* dpi - Разрешение рисунка в точках на дюйм (по умолчанию 100).
* facecolor - Цвет фона (по умолчанию 'white').
* edgecolor - Цвет границы(по умолчанию 'white').
* frameon - Если значение False, отключает рисование рамки фигуры (по умолчанию True).
* [и др параметры](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html).

Метод Figure.add_subplot() - Добавить подграфик к рисунку

Три способа вызова (по умолчанию передается (1, 1, 1)):
* Figure.add_subplot(<k строк>, <k столбцов>, <индекс позиции>>)
* Figure.add_subplot(<трезначное число>) - аналогично прошлому способу, но здесь три параметра объединяются в один
* Figure.add_subplot(GridSpec)

```python
# Создание нового рисунка с последующим добавлением подграфиков
fig = plt.figure()
ax1 = fig.add_subplot()
```

```python
# Построение графика с заданной структурой и размером
fig = plt.figure(figsize=(4, 4))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
```

```python
# Использование других параметров figure()
fig = plt.figure(dpi=70, facecolor='orange')
ax1 = fig.add_subplot()
```

```python
# Построение графика с более сложной структурой
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1) # нарисовать график на первом месте, при учете, что 1 строка, 2 столбца (рисуется левый)
ax2 = fig.add_subplot(2, 2, 2) # нарисовать график на 2 месте, при учете, что 2 строки, 2 столбца
ax3 = fig.add_subplot(2, 2, 4) # нарисовать график на 4 месте, при учете, что 2 строки, 2 столбца
```

**GridSpec** - Макет сетки для размещения подграфиков внутри рисунка.
* nrows, ncols - Количество строк и столбцов сетки.
* fig - полотно, к которому будет прикреплен GridSpec

Для его использования необходимо импортировать *import matplotlib.gridspec as gridspec*

Метод Figure.add_gridspec - альтернативный метод создания GridSpec сразу для конкретной Figure. На вход принимает nrow, ncols.

```python
import matplotlib.gridspec as gridspec

# Cоздание с помощью конструктора
fig = plt.figure()
gs = gridspec.GridSpec(1, 2, figure=fig)
fig.add_subplot(gs[0, 0])
fig.add_subplot(gs[0, 1])
```

```python
# Cоздание с помощью метода add_gridspec
fig = plt.figure()
gs = fig.add_gridspec(1, 2)
fig.add_subplot(gs[0, 0])
fig.add_subplot(gs[0, 1])
```

```python
# Более сложная структура
fig = plt.figure()
gs = fig.add_gridspec(3, 3)

#Добавим на каждый график число, чтобы понимать соответствие строки и подграфика
fig.add_subplot(gs[0:2, 0:2]).text(0.5, 0.5, "1", fontsize=20) 
fig.add_subplot(gs[0, 2]).text(0.5, 0.5, "2", fontsize=20)
fig.add_subplot(gs[2, 0:2]).text(0.5, 0.5, "3", fontsize=20)
fig.add_subplot(gs[1, 2]).text(0.5, 0.5, "4", fontsize=20)
fig.add_subplot(gs[2, 2]).text(0.5, 0.5, "5", fontsize=20)
```

При работе в Jupyter необязательно каждый раз создавать полотно и подграфики, а затем оперировать с ними. Все методы Figure и Axes продублированы в matplotlib.pyplot. При вызове метода, который соответствует одному из двух типов данных, matplotlib автоматически создаст соответствующий объект, если он еще не был создан. Есть некоторые отличия в сигнатурах методов, в основном они заключаются в том, что для Figure и Axes нужно писать set/get/add_method, а для pyplot достаточно просто .method

```python
plt.figure(figsize=(4, 4)) # был создан figure 
plt.subplot(121) # В среду записан, созданный на прошлом шаге подграфик, метод будет вызван над ним
plt.text(0.5, 0.5, "1", fontsize=20) # Отрисовка текста на подграфике, который был создан на прошлом шаге
plt.subplot(122)
plt.text(0.5, 0.5, "2", fontsize=20)
```

# Основные элементы графика Matplotlib

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot()

fig.suptitle('Заголовок фигуры') # Выставление заголовка фигуры
ax.set_title("Заголовок подграфика") # Выставление заголовка подграфика
ax.set_xlabel('X') # Выставление названия оси по X
ax.set_ylabel('Y') # Выставление названия оси по Y
ax.set_xticks([10, 20, 30, 40, 50, 60]) # Значения для отрисовки метки по оси X
ax.set_yticks([1, 2, 3, 4, 5, 6]) # Значения для отрисовки метки по оси Y
ax
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot()
ax.set_xticklabels(['a', 'b', 'c', 'd', 'e', 'f']) # Метки по оси X (просто строки, не значения)
ax.set_yticklabels(['A', 'B', 'C', 'D', 'E', 'F']) # Метки по оси Y (просто строки, не значения)
ax.plot()
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot()
ax.set_xlim([-10, 30]) # Ограничение графика по оси X
ax.set_ylim([-5, 10]) # Ограничение графика по оси Y
ax.grid() # Отрисовка решетки
```

# Точечный график

Точечный график используется для визуализации двух переменных, показывая зависимость между ними. Он позволяет наглядно оценить, как одна переменная влияет на другую, а также выявить возможные выбросы и аномалии в данных.

### Точечный график в Matplotlib

plt.scatter - метод Matplotlib для отрисовки точечного графика.
* x, y - списки с позициями по x и по y. Точки берутся по позициям (x\[i\], y\[i\])
* s - опциональное поле с размерами маркеров. Может быть числом или списком с сопоставлением для каждой точки
* с - цвет маркера
* marker - тип маркера. По умолчанию точечный график изображается точками, но маркер может быть изменен.
* alpha - прозрачность маркеров
* linewidths - ширина линии краев маркера
* edgecolor - Цвет края маркера
* [и другие ](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

##### Допустимые маркеры

|Обозначение маркера|Описание маркера|
|--|--|
|'.'|Точка (кружок)|
|','|Пиксель|
|'o'|Круг (больше чем точка)|
|'v'|Треугольник вниз|
|'^'|Треугольник вверх|
|'s'|Квадрат|
|'p'|Пятиугольник|
|'P'|Плюс|
|'x'|Крестик|

Маркеров огромное множество, остальные можно посмотреть [тут](https://matplotlib.org/stable/api/markers_api.html).

```python
# Простая отрисовка точек
x = np.random.sample(20, ) * 10
y = np.random.sample(20, ) * 10
plt.scatter(x, y)
```

```python
# Отрисовка точечного графика с размерами
s = np.random.sample(20, ) * 500
plt.scatter(x, y, s=s)
```

```python
# Использование других параметров scatter
plt.scatter(x, y, c='orange', s=150, marker='s', linewidths=2, edgecolors='blue')
```

# Линейный график

### Линейный график в Matplotlib

plt.plot - метод Matplotlib для отрисовки линейного графика
* x, y - списки с значениями функции по x и по y.
* alpha - прозрачность линии
* color - цвет линии
* label - лейбл для последующей отрисовки легенды
* linestyle - тип линии (например, штриховая, штрих-пунктирная)
* linewidth - ширина линии
* marker - тип маркера, который будет ставиться в каждой точке графика (x, y)
* markeredgecolor - цвет обводящей линии вокруг маркера
* markeredgewidth - ширина обводящей линии вокруг маркера
* markerfacecolor - цвет маркера
* markersize - размер маркера
* antialiased или aa - делать ли линию сглаженной (по умолчанию True)

```python
x = np.arange(0, 20, 0.2)
y = [np.sin(x) for x in x]
plt.plot(x, y)
```

##### Стили линий

|Обозначение|Стиль линии|
|--|--|
|-|Непрерывная|
|--|Штриховая линия|
|-.|Штрихпунктирная|
|:|Пунктирная|

##### Множественные графики Matplotlib

**Для построения нескольких графиков на одном подграфике достаточно несколько раз вызвать функцию отрисовки графика для объекта данных данного подграфика.**

```python
x = np.arange(0, 20, 0.2)
y1 = [np.sin(x) for x in x]
y2 = [np.cos(x) for x in x]
plt.plot(x, y1)
plt.plot(x, y2)
```

##### Легенда графика в Matplotlib

Для множественных графиков в Matplolib удобно иметь на графике легенду. Легенда может сопоставлять тип линии/цвет заливки/тип точки и график, который она характеризует (описание графика берется из переданного параметра label).

```python
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.legend()
```

```python
# Можно указать какие лейблы должны быть в легенде
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.legend(['cos'])
```

Рассмотрим еще один пример для линейных графиков со всей информацией, которую мы теперь знаем

```python
x = np.arange(0, 20, 0.4)
plt.figure(figsize=(10, 6))
y1 = [np.sin(x) for x in x]
y2 = [np.cos(x) for x in x]
y3 = [np.cos(2*x) for x in x]
plt.plot(x, y1, label='sin(x)', color='red', marker='s', markersize='7', markeredgecolor='blue', markeredgewidth='2')
plt.plot(x, y2, label='cos(x)', color='gold', marker='o', linestyle='--', linewidth='4', markerfacecolor='green')
plt.plot(x, y3, label='cos(2x)', aa=False, alpha=0.5)
plt.legend()
```

# Площадная диаграмма

Площадные диаграммы используются для визуализации данных, показывающих соотношение значений между различными категориями. Площадные диаграммы могут быть полезны для сравнения вклада каждой категории в общую сумму. Помимо этого площадные диаграммы могут выполнять роль линейной функции, но с заливкой цветом под графиком.

### Площадная диаграмма в Matplotlib

Метод plt.fill_between - заполнение пространства между двумя горизонтальными кривыми.
* x - координаты графика по оси X.
* y1 - координаты первой кривой по оси Y.
* y2 - координаты второй кривой по оси Y (по умолчанию - ось абсцисс).
* where - логическое выражение, определяющее, при каких условиях нужно рисовать график. Может быть полезно, если заполнение нужно только в определенной области значений X.

Есть аналогичный метод для вертикальных кривых - plt.fill_betweenx

```python
# Заполнение графика, где вторая линия - ось абсцисс
x = np.arange(0, 20, 0.2)
y1 = [np.sin(x) for x in x]
plt.fill_between(x, y1)
```

```python
# Заполнение графика, где вторая линия - тоже кривая
x = np.arange(0, 5, 0.2)
y1 = [np.sin(x) for x in x]
y2 = [np.cos(x) for x in x]
plt.fill_between(x, y1, y2)
```

```python
# Заполнение графика, где вторая линия - тоже кривая, с ограничением на X
x = np.arange(0, 5, 0.2)
y1 = [np.sin(x) for x in x]
y2 = [np.cos(x) for x in x]
plt.fill_between(x, y1, y2, where=(x > 1) & (x < 2))
# Добавим также выделение границ
plt.plot(x, y1)
plt.plot(x, y2)
```

* x - координаты графика по оси X размера N.
* y - список списков с значениями графика по оси Y. Вложенный список также должен иметь размер N.
* colors - список цветов для каждого уровня площадной диаграммы.
* labels - список лейблов для каждого уровня площадной диаграммы.

# Столбчатые и линейчатые диаграммы

Столбчатая и линейчатая диаграмма показывает относительно распределение значений по какому-либо критерию.

Часто эти диаграммы могут быть использованы для отображения изменения во времени. 

Как и в случае площадной диаграммы, столбчатые и линейчатые могут отображать накопительный эффект.

**Гистограмма** - это частный случай столбчатой диаграммы, которая отображает распределение непрерывных данных. Непрерывные данные - это данные, которые могут принимать любые значения в некотором интервале. Например, средняя продолжительность жизни/температура/вес.

В отличие от линейчатой и столбчатой диаграмм, ось X не делится на взаимоисключающие категории, это сплошная шкала, которая делится на равные интервалы, а по оси Y откладывается количество данных, попадающих в этот интервал. 

### Столбчатые и линейчатые диаграммы в Matplotlib

##### Столбчатая диаграмма

Метод plt.bar - построение столбчатой диаграммы.
* x - координаты столбцов по оси X.
* height - список высот столбцов.
* width - ширина столбцов (по умолчанию 0.8).
* bottom - нижняя кордината столбцов диаграммы, может быть списком (по умолчанию 0, то есть ось абсцисс).
* align - выравнивание столбцов по координатам X. Если 'center', то координаты располагаются по центру столбца, если 'edge'- выравнивание по левому краю столбца (по умолчанию 'center').
* color - цвет или список цветов для столбцов диаграммы.
* edgecolor - цвет краев столбца
* linewidth - ширина краев столбца (по умолчанию 0)
* label - лейбл для графика

```python
# Отрисовка по умолчанию
x = [1, 2, 3, 4, 5, 6, 7]
height = [4, 3, 5, 6, 8, 3, 9]
plt.bar(x, height)
```

```python
# Использование параметров
plt.bar(
    x, 
    height, 
    width=0.5,
    color=['red', 'green', 'blue', 'gold', 'orange', 'purple', 'silver'],
    edgecolor=list(reversed(['red', 'green', 'blue', 'gold', 'orange', 'purple', 'silver'])),
    linewidth=2)
```

```python
# Построение столбчатой диаграммы с несколькими столбцами
x = [1, 2, 3, 4, 5, 6, 7]
height1 = [4, 3, 5, 6, 8, 3, 9]
height2 = [2, 3, 1, 3, 4, 2, 2]
# Отступ вправо/влево равен ширине деленной на 2
plt.bar(np.array(x) - 0.2, height1, width=0.4)
plt.bar(np.array(x) + 0.2, height2, width=0.4)
```

```python
# Построение накопленной столбчатой диаграммы
plt.bar(x, height1)
plt.bar(x, height2, bottom=height1)
```

# Круговая диаграмма

Круговая диаграмма используется для визуализации данных, которые представляют собой доли от целого. Она помогает быстрее и нагляднее сравнить относительные размеры различных частей целого.

### Круговая диаграмма в Matplotlib

Метод plt.pie
* x - размер долей круговой диаграммы.
* explode - список значений размера len(x), который определяет смещение каждой доли диаграммы от центра.
* labels - список лейблов для каждой доли диаграммы.
* autopct - формат текстовой метки для отображения чисел, характеризующие доли диаграммы (по умолчанию None - без отображения).
* pctdistance - расстояние от центра диаграммы, на котором будут отображены текстовые метки из autopct (по умолчанию 0.6).
* colors - список цветов для каждой диаграммы.
* labeldistance - расстояние от центра диаграммы, на котором будут отображены лейблы. Если None, то лейблы не отображаются (по умолчанию 1.1).
* shadow - если True, то отображается тень для диаграммы (по умолчанию False).
* startangle - стартовый угол отображения долей диаграммы (по умолчанию 0).
* radius - радиус круговой диаграммы.
* counterclock - определяет направление отрисовки круговой диаграммы. Если True, то против часовой (по умолчанию True).
* wedgeprops - словарь аргументов и значений, которые применяются к каждой доли диаграммы.
  * linewidth - ширина обводной линии (по умолчанию 1).
  * edgecolor - цвет обводной линии.
  * linestyle - тип обводной линии, аналогично линейному графику.
* textprops - словарь аргументов и значений, которые применяются ко всем текстовым объектам графика.
  * color - цвет текста.
  * fontstyle - шрифт.
  * fontsize - размер текста.
* center - координаты центра диаграммы (по умолчанию (0, 0)).
* rotatelabels - если True, то лейблы будут повернуты, соответственно углам долей.

```python
# Отрисовка по умолчанию
x = [2, 1, 6, 4, 8]
plt.pie(x)
```

```python
# Используем параметры
plt.pie(x, 
        labels=['red', 'green', 'blue', 'orange', 'gray'], 
        colors=['red', 'green', 'blue', 'orange', 'gray'])
plt.plot()
```

```python
# Параметры labeldistance, shadow, explode, rotatelabels
plt.pie(x, 
        labels=['red', 'green', 'blue', 'orange', 'gray'], 
        colors=['red', 'green', 'blue', 'orange', 'gray'],
        labeldistance=1.4,
        shadow=True,
        explode=[0, 0, 0.3, 0.2, 0],
        rotatelabels=True)
plt.plot()
```

```python
# Параметры autopct, pctdistance, counterclock, startangle
plt.pie(x, 
        autopct='%d',
        pctdistance=0.9,
        counterclock=True,
        startangle=90)
plt.plot()
```

```python
# Параметры counterclock, wedgeprops, textprops
plt.pie(x, 
        labels=['red', 'green', 'blue', 'orange', 'gray'], 
        colors=['red', 'green', 'blue', 'orange', 'gray'],
        counterclock=False,
        wedgeprops={'linewidth': 3, 
                    'edgecolor': 'black',
                    'linestyle': '--'},
        textprops={'color': 'purple',
                   'fontsize': 20})
plt.plot()
```

# Seaborn.pairplot

Метод Seaborn.pairplot предоставляет возможность построить графики, отражающие попарные отношения в наборе данных.
* data - DataFrame, по которому будут строиться подграфики.

* palette - набор соответствий для отображения переменной hue.
* vars - набор столбцов, которые необходимо включить в график, если не указано, то возьмутся все с числовым типом данных.
* kind - тип графика для отрисовки соответствий. Один из {'scatter', 'kde', 'hist', 'reg'}
* diag_kind - тип графика для отрисовки по диагонали. Один из {'auto', 'hist', 'kde'}
* markers - маркеры для использования при типе графика - scatter
* corner - если True, подграфики не добавляются к верхнему треугольнику сетки, делая этот участок “угловым”.
* dropna - удалить недостающие значения из данных перед построением графика

# Тепловая карта Seaborn

Тепловая карта помогает пользователю быстро установить какие-то закономерности и тенденции, которые могут быть не очевидны при простом просмотре числовых значений. 

Метод Seaborn.heatmap
* data - список данных n*m или DataFrame широкой формы
* vmin, vmax - значения минимального и максимального значения цветовой карты. Если не указывается, то выводятся из данных.
* cmap - цветовая карта (Цветовая карта представляет собой систематизированный набор цветов, например в графиках он используется, чтобы показать, какое значение больше/меньше. Подробней про цветовые карты и их виды в Matplotlib [здесь](https://matplotlib.org/stable/tutorials/colors/colormaps.html)).
* center - значение, по которому центрируется цветовая карта.
* annot - если True, то каждая ячейка тепловой карты будет аннотирована соответствующим числовым значением.
* annot_kws - словарь аргументов и значений, которые применяются к аннотации, например linewidth, fontsize.'
* linewidths - ширина линий разделяющая ячейки между собой.
* linecolor - цвет разделяющих линий.
* cbar - если True, то выводит цветовую карту (по умолчанию True).
* xticklabels, yticklabels - список лейблов для меток по оси X и Y, соответственно (по умолчанию Seaborn вычисляет метки автоматически).
* ax - экземпляр класса Axes, на котором будет отображен график.

```python
# Для DataFrame широкой формы
df = pd.DataFrame({'team': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
 'player': [1, 2, 3, 4, 1, 2, 3, 4],
 'points': [11, 8, 10, 6, 12, 5, 9, 4]})
df = df.pivot('team', 'player', 'points')
sns.heatmap(df)
```
